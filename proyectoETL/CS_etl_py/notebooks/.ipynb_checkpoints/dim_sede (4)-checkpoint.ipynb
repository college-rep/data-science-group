```python
from datetime import date

import numpy as np
import pandas as pd
import yaml
from sqlalchemy import create_engine
```

# database connections 

```python
with open('../config.yml', 'r') as f:
    config = yaml.safe_load(f)
    config_co = config['CO_SA']
    config_etl = config['ETL_PRO']

url_co = (f"{config_co['drivername']}://{config_co['user']}:{config_co['password']}@{config_co['host']}:"
          f"{config_co['port']}/{config_co['dbname']}")
url_etl = (f"{config_etl['drivername']}://{config_etl['user']}:{config_etl['password']}@{config_etl['host']}:"
           f"{config_etl['port']}/{config_etl['dbname']}")
co_sa = create_engine(url_co)
etl_conn = create_engine(url_etl)
```

# Extract

Sede no trae el nombre del departamento directo, se necesita el join `sede -> ciudad -> departamento`.

```python
dim_sede = pd.read_sql_query('''
    SELECT
        s.sede_id,
        s.nombre   AS nombre_sede,
        s.direccion,
        c.nombre   AS ciudad,
        d.nombre   AS departamento,
        s.cliente_id
    FROM sede s
    LEFT JOIN ciudad c       ON s.ciudad_id = c.ciudad_id
    LEFT JOIN departamento d ON c.departamento_id = d.departamento_id
''', co_sa)
dim_sede.info()
dim_sede
```

# Transformations

Los `LEFT JOIN` pueden dejar `ciudad` o `departamento` en nulo cuando la sede no tiene la referencia completa; se normalizan esos vacios. `cliente_id` se conserva como atributo extra para responder la pregunta de sedes por cliente.

```python
dim_sede.replace({np.nan: 'no aplica', ' ': 'no aplica', '': 'no aplica'}, inplace=True)
dim_sede['saved'] = date.today()
dim_sede
```

# Alinear nombres de columnas al diagrama
ID_Sede (PK) -> indice al cargar | Nombre_Sede | Direccion | Ciudad | Departamento

`cliente_id` no esta en el diagrama, se deja como atributo extra y no estorba.

```python
dim_sede.rename(columns={
    'sede_id': 'id_sede',
    'cliente_id': 'id_cliente',
}, inplace=True)
dim_sede
```

# load

```python
# NOTA: if_exists='append' porque la tabla se crea por DDL (ver sqlscripts.yml) con su PK real.
# 'replace' borraria esa tabla y la reemplazaria sin restricciones. Antes de correr este notebook,
# asegurate que el DDL ya creo la tabla dim_sede (o ejecuta el script de sqlscripts.yml manualmente).
dim_sede.to_sql('dim_sede', etl_conn, if_exists='append', index_label='key_dim_sede')
```
