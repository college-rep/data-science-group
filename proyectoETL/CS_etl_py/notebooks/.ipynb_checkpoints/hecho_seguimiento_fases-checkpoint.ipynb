```python
from datetime import date

import numpy as np
import pandas as pd
import yaml
from sqlalchemy import create_engine
## preguntas que responde este hecho (Seguimiento_Fases)
###### 3. A que hora los mensajeros estan mas ocupados
###### 5. Mensajeros mas eficientes (los que mas servicios prestan)
###### 7. Tiempo promedio de entrega desde que se solicita hasta que se cierra
###### 8. Tiempos de espera por cada fase / en que fase hay mas demoras
```

# database connections 

```python
with open('../config.yml', 'r') as f:
    config = yaml.safe_load(f)
    config_co = config['CO_SA']
    config_etl = config['ETL_PRO']

url_co = (f"{config_co['drivername']}://{config_co['user']}:{config_co['password']}@{config_co['host']}:"
          f"{config_co['port']}/{config_co['dbname']}")
url_etl = (f"{config_etl['drivername']}://{config_etl['user']}:{config_etl['password']}@{config_etl['host']}:"
           f"{config_etl['port']}/{config_etl['dbname']}")
co_sa = create_engine(url_co)
etl_conn = create_engine(url_etl)
```

# Extract

La tabla de fases `mensajeria_estadosservicio` solo trae `fecha, hora, estado_id, servicio_id`. El cliente y el mensajero salen de `mensajeria_servicio`; la sede se resuelve desde `dim_sede` (cliente -> sede), igual que en `hecho_servicio`.

```python
tabla_fases = pd.read_sql_table('mensajeria_estadosservicio', co_sa)
servicios   = pd.read_sql_table('mensajeria_servicio', co_sa)
dim_sede    = pd.read_sql_table('dim_sede', etl_conn)
dim_fecha   = pd.read_sql_table('dim_fecha', etl_conn)
tabla_fases.info()
```

# Transformations

**Calculo del tiempo de espera por fase:** se ordenan los estados de cada servicio por fecha+hora y se mide la diferencia hasta el **siguiente** estado de ese mismo servicio. Esa diferencia es el tiempo que el servicio paso en cada fase. La ultima fase no tiene siguiente, asi que su `tiempo_espera_min` queda nulo.

Se descartan los registros de prueba (`es_prueba = True`).

```python
hecho = tabla_fases[tabla_fases['es_prueba'] == False].copy()
hecho = hecho[['id', 'fecha', 'hora', 'estado_id', 'servicio_id']]

# timestamp combinando fecha + hora para poder ordenar y restar
hecho['ts'] = pd.to_datetime(hecho['fecha'].astype(str) + ' ' + hecho['hora'].astype(str))

# tiempo en cada fase = diferencia hasta el siguiente estado del MISMO servicio
hecho = hecho.sort_values(['servicio_id', 'ts'])
hecho['ts_siguiente'] = hecho.groupby('servicio_id')['ts'].shift(-1)
hecho['tiempo_espera_min'] = (hecho['ts_siguiente'] - hecho['ts']).dt.total_seconds() / 60

hecho.info()
```

```python
# Traer cliente y mensajero desde mensajeria_servicio (join por servicio)
servicios_min = servicios[['id', 'cliente_id', 'mensajero_id']].rename(columns={'id': 'servicio_id'})
hecho = hecho.merge(servicios_min, on='servicio_id', how='left')

# Resolver la sede a partir del cliente (misma logica que hecho_servicio)
sede_por_cliente = dim_sede[['id_cliente', 'id_sede']].drop_duplicates(subset='id_cliente')
hecho = hecho.merge(sede_por_cliente, left_on='cliente_id', right_on='id_cliente', how='left')

# id_hora = minuto del dia (coincide con key_dim_hora). id_fecha = key_dim_fecha por la fecha
hecho['id_hora'] = hecho['ts'].dt.hour * 60 + hecho['ts'].dt.minute
dim_fecha_min = dim_fecha[['key_dim_fecha', 'fecha_completa']].copy()
dim_fecha_min['fecha_completa'] = pd.to_datetime(dim_fecha_min['fecha_completa']).dt.date
hecho = hecho.merge(dim_fecha_min, left_on='fecha', right_on='fecha_completa', how='left')
hecho.rename(columns={'key_dim_fecha': 'id_fecha'}, inplace=True)
hecho.info()
```

# Alinear nombres de columnas al diagrama
ID_Fecha | ID_Hora | ID_Cliente | ID_Sede | ID_Mensajero | ID_Estado | id_servicio | Tiempo_Espera_Min

```python
hecho.rename(columns={
    'cliente_id': 'id_cliente',
    'mensajero_id': 'id_mensajero',
    'estado_id': 'id_estado',
}, inplace=True)

hecho_seguimiento_fases = hecho[[
    'id_fecha', 'id_hora', 'id_cliente', 'id_sede',
    'id_mensajero', 'id_estado', 'servicio_id', 'tiempo_espera_min'
]].rename(columns={'servicio_id': 'id_servicio'})

hecho_seguimiento_fases.info()
hecho_seguimiento_fases.head()
```

# load

```python
hecho_seguimiento_fases.to_sql('hecho_seguimiento_fases', etl_conn, if_exists='replace', index_label='key_hecho_seguimiento_fases')
```
